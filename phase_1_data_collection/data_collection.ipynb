## <p style="text-align: center;">**Phần 2 - Nạp bộ dữ liệu** <a name="answer-the-question"></a></p> 

Import dữ liệu về phim lẻ & phim truyền hình từ 3 nền tảng Netflix, Amazon Prime và HBO Max đã tải trên Kaggle.

##### **Bước 1: Import thư viện cần thiết**

```python
import pandas as pd
```

##### **Bước 2: Đọc vào pandas các bộ dữ liệu từ 3 nền tảng.**
Mỗi bộ dữ liệu sẽ có 2 tập tin khác nhau, có vai trò lần lượt như sau:
-  Tập tin có tên là titles dùng để ghi lại thông tin chính của bộ phim.
-  Tập tin có tên là credits dùng để ghi lại thông tin của đạo diễn và diễn viên của phim.

Đọc dữ liệu từ nền tảng phân phối phim Amazon:

```python
df_amazon_titles = pd.read_csv('../data/amazon/titles.csv')
df_amazon_credits = pd.read_csv('../data/amazon/credits.csv')
```

Đọc dữ liệu từ nền tảng phân phối phim HBO:

```python
df_hbo_titles = pd.read_csv('../data/hbo/titles.csv')
df_hbo_credits = pd.read_csv('../data/hbo/credits.csv')
```

Đọc dữ liệu từ nền tảng phân phối phim Netflix.

```python
df_netflix_titles = pd.read_csv('../data/netflix/titles.csv')
df_netflix_credits = pd.read_csv('../data/netflix/credits.csv')
```

##### **Bước 3: Nối các bộ dữ liệu thành một bảng dữ liệu chung cho cả 3 nền tảng:**
Vì các bộ dữ liệu credits cũng như titles có các thuộc tính giống nhau nên có thể nối được thành một bảng lớn tương ứng.

```python
df_titles_raw = pd.concat([df_amazon_titles, df_hbo_titles, df_netflix_titles], axis=0)
df_credits_raw = pd.concat([df_amazon_credits, df_hbo_credits, df_netflix_credits], axis=0)
```

In ra để kiểm tra thông tin phim đã được nối bảng hay chưa:

```python
df_titles_raw
```

In ra để kiểm tra thông tin diễn viên, đạo diễn của bộ phim được nối lại từ 3 bộ dữ liệu:

```python
df_credits_raw
```
